```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/RPGraciotti/BootCampAlura/main/Modulo_1/A191953189_28_143_208.csv"
df = pd.read_csv(url, 
                 encoding = "ISO-8859-1",
                 skiprows = 3, sep = ";", skipfooter = 12,
                 thousands = ".", decimal = ",", engine = "python")
df.head()
```

```python
df.mean()
```

```python
pd.options.display.float_format = "{:.2f}".format
```

```python
df.mean()
```

```python
df.loc[:,'2008/Ago'].mean()
```

```python
df.plot(x = "Unidade da Federação", y = "2008/Ago", kind = "bar", 
        figsize = (10,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
ax = df.plot(x = "Unidade da Federação", y = "2008/Ago", kind = "bar", 
        figsize = (10,7)) # controle dos eixos
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por Unidade da Federação")
plt.show()
```

## 2) Faça a mesma análise realizada em aula, porém para o mês mais recente:

### Identificar o mês mais recente do conjunto de dados:

```python
df.shape # 27 linhas e 161 colunas
df.columns # Ultimo mês = 2021/Mar
print(df.loc[:,"2021/Mar"]) # por label
print()
print(df.iloc[:,159]) # por posição
```

### Plotando esse mês

#### Por label

```python
ax = df.plot(x = "Unidade da Federação", y = "2021/Mar", kind = "bar", 
        figsize = (10,7)) # controle dos eixos
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor total (em reais) processado em atendimentos hospitalares por UF no Março de 2021")
plt.show()
```
